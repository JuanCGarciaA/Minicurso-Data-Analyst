# Listas en Python

Las listas en Python son objetos que pueden contener cualquier tipo de elemento, por ejemplo, enteros, cadenas de caracteres, otras listas e incluso DataFrames.

No es necesario que todos los elementos de una lista sean del mismo tipo de variable, por ejemplo, podríamos tener la siguiente lista con elementos enteros y cadenas de caracteres:

```python
[1, 2, "a"]
```

Es muy común usar bucles for para recorrer todos los elementos de una lista y realizar alguna operación:

```python
for element in my_list:
  # do something with 'element'
```

# Ejercicio 1

Dada una lista, completa la siguiente función para que rellene con valores nulos si un elemento de dicha lista es de tipo string.

Tu función debería de regresar una lista con valores `None` en caso de que originalmente hayan sido elementos de tipo string.

Tip: usa la función `isinstance(element, str)` para determinar si un elemento es de tipo string.

Tu código debería de pasar las pruebas escritas más abajo.

````python
def nullify_strings(test_list):
  listAux = []
  for element in test_list:
    if isinstance(element,str):
      listAux.append(None)
    else:
      listAux.append(element)
  return listAux
````

````python
list1 = ["a", 2, 3]
list2 = [4, 5, 6, 7]
list3 = ["a", "b", "c"]
list4 = [1, "b", 5.0, [3, "d", 4.0]]

print(nullify_strings(list1) == [None, 2, 3])
print(nullify_strings(list2) == [4, 5, 6, 7])
print(nullify_strings(list3) == [None, None, None])
print(nullify_strings(list4) == [1, None, 5.0, [3, "d", 4.0]])
````

# Ejercicio 2

Sin usar la función `min()` (o `max()`) y usando solo ciclos `for`, determina el valor mínimo dentro de una lista.

Tu función debería de regresar el elemento más pequeño de la lista.

Tu código debería de pasar las pruebas escritas más abajo.

````python
def custom_min(test_list):
    n = len(test_list)
    if n == 0:
        return None

    for i in range(n-1):
        min_index = i
        for j in range(i+1, n):
            if test_list[j] < test_list[min_index]:
                min_index = j
        test_list[i], test_list[min_index] = test_list[min_index], test_list[i]

    return test_list[0]
````

````python
list1 = [1, 2, 3]
list2 = [7, 6, 5]
list3 = [2, 4, 5, 2]
list4 = [0]

print(custom_min(list1) == 1)
print(custom_min(list2) == 5)
print(custom_min(list3) == 2)
print(custom_min(list4) == 0)
````

# Ejercicio 3

Escribe una función que calcule cuántos elementos duplicados existen en una lista.

Tu función debería de regresar un número entero correspondiente al número de elementos duplicados en la lista original.

Tip: usa la función `set` para obtener los valores únicos de una lista.

Tu código debería de pasar las pruebas escritas más abajo.

````python
def count_duplicates(test_list):
    vals = set(test_list)
    duplic = 0

    for valor in vals:
        if test_list.count(valor) > 1:
            duplic += 1

    return duplic
````

````python
list1 = [1, 1, 2]
list2 = [1, 2, 2, 2, 3, 3, 4]
list3 = [1, 2, 3]

print(count_duplicates(list1) == 1)
print(count_duplicates(list2) == 2)
print(count_duplicates(list3) == 0)
````

# Ejercicio 4

Calcula la distancia mínima que existe entre todos los elementos de una lista.

La función debería de regresar un número correspondiente a la distancia mínima entre todos elementos de la lista.

Tu código debería de pasar las pruebas escritas más abajo.

````python
def minimum_distance(test_list):
    ordered_list = sorted(test_list)
    initial_min_dist = ordered_list[1] - ordered_list[0]

    for i in range(len(ordered_list) - 1):
        dist = ordered_list[i+1] - ordered_list[i]
        if dist < initial_min_dist:
            initial_min_dist = dist

    return initial_min_dist
````

````python
list1 = [1, 5, 2]
list2 = [1, 4, 7, 11]
list3 = [1, 5]
list4 = [1, 5, 9, 10]

print(minimum_distance(list1) == 1)
print(minimum_distance(list2) == 3)
print(minimum_distance(list3) == 4)
print(minimum_distance(list4) == 1)
````

# Ejercicio 5

Replica la función `sorted(list)` usando solo ciclos for, es decir, construye una función que ordene de manera ascendente una lista.

Tu función debería de regresar la misma lista en orden ascendente.

Tu código debería de pasar las pruebas escritas más abajo.

````python
def custom_sort(test_list):
  n = len(test_list)
  if n == 0:
    return None

  for i in range(n-1):
    min_index = i
    for j in range(i+1, n):
      if test_list[j] < test_list[min_index]:
        min_index = j
      test_list[i], test_list[min_index] = test_list[min_index], test_list[i]
  return test_list
````

````python
list1 = [1, 1, 2, 5]
list2 = [10, 10, 7, 4, 3]
list3 = [5]

print(custom_sort(list1) == [1, 1, 2, 5])
print(custom_sort(list2) == [3, 4, 7, 10, 10])
print(custom_sort(list3) == [5])
````
