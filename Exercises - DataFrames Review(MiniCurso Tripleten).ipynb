# Ejercicio 1

Escribe una función que verifique si dos DataFrames tienen el mismo número de renglones. La función debería de arrojar un valor booleano indicando si los dos DataFrames tienen el mismo número de renglones.

Tu código debería de pasar las pruebas escritas más abajo.

```python
def same_number_of_rows(df1, df2):
  if (len(df1.index) == len(df2.index)):
    return True
  else:
    return False
```

```python
df11 = pd.DataFrame(data={"a": [1, 2, 3], "b": [4, 5, 6]})
df12 = pd.DataFrame(data={"a": [1, 2], "b": [4, 5]})
df21 = pd.DataFrame(data={"a": [10, 20, 30], "b": [40, 50, 60]})
df22 = pd.DataFrame(data={"a": [100, 200, 300], "b": [400, 500, 600]})

print(same_number_of_rows(df11, df12) == False)
print(same_number_of_rows(df21, df22) == True)
```

# Ejercicio 2

Escribe una función que determine si el nombre de las columnas es igual entre dos DataFrames.

Tu función debería de arrojar un valor booleano indicando si dos DataFrames contienen las mismas columnas.

Tu código debería de pasar las pruebas escritas más abajo.

```python
def have_same_columns(df1, df2):
  if(sorted(df1.columns.tolist()) == sorted(df2.columns.tolist())):
    return True
  else:
    return False
```

```python
df11 = pd.DataFrame(data={"a": [1, 2, 3], "b": [4, 5, 6]})
df12 = pd.DataFrame(data={"a": [1, 2], "b": [4, 5]})
df21 = pd.DataFrame(data={"b": [10, 20, 30], "a": [40, 50, 60]})
df22 = pd.DataFrame(data={"a": [100, 200, 300], "b": [400, 500, 600]})
df31 = pd.DataFrame(data={"bb": [10, 20, 30], "a": [40, 50, 60]})
df32 = pd.DataFrame(data={"a": [100, 200, 300], "b": [400, 500, 600]})

print(have_same_columns(df11, df12) == True)
print(have_same_columns(df21, df22) == True)
print(have_same_columns(df31, df32) == False)
```

Tu función debería de arrojar un string correspondiente a la columna con mayor número de valores vacíos.

Tu código debería de pasar las pruebas escritas más abajo.

```python
def col_with_most_nulls(df):
  dic= {df.columns.values[0] :0 , df.columns.values[1] : 0 }
  for elem in df.columns.values:
    cnt = 0
    for unit in df[elem].values:
      if math.isnan(unit):
        cnt += 1
    dic[elem] = cnt
  for it in dic.values():
    cnt = 0
    if it == 0:
      cnt += 0
      if cnt == 0:
        return None
  return min(dic)
```

```python
df1 = pd.DataFrame(data={"a": [1, None, None], "b": [4, 5, None]})
df2 = pd.DataFrame(data={"a": [1, 2], "b": [4, 5]})
df3 = pd.DataFrame(data={"abd": [1, 2, None], "abc": [4, 5, None]})

print(col_with_most_nulls(df1) == "a")
print(col_with_most_nulls(df2) == None)
print(col_with_most_nulls(df3) == "abc")
```
